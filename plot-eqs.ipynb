```python
car_sales.Price = car_sales.Price.astype(str).str.replace('\$|\.00|,', "", regex=True).astype(int)
```

```python
car_sales['Date'] = pd.date_range("1/1/2018", periods=len(car_sales))
```

```python
car_sales['Revenue'] = car_sales["Price"].cumsum()
```

```python
car_sales.plot(x='Date', y='Revenue');
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease.age.plot.hist(bins=100)
```

```python
over_50 = heart_disease[heart_disease['age'] > 50];
over_50.plot(kind="scatter", x="age", y="chol", c="target", figsize=(10, 5));
```

```python
fig, ax = plt.subplots(figsize=(10, 5), dpi=72)

ax.set(ylabel='Cholestrol', 
       xlabel='Age',
       title="Cholestrol vs Age");

scatter = ax.scatter(
    data=over_50,
    x='age', 
    y='chol', 
    c='target',
);

ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(y=over_50.chol.mean(), linestyle=":");
```

```python
fig, (ax0, ax1) = plt.subplots(figsize=(10, 10), nrows=2, sharex=True)

fig.suptitle("Heart Disease analysis", fontsize=20, fontweight='bold')

# Cholestrol scatter plot
ax0.set(ylabel='Cholestrol', 
       title="Cholestrol vs Age");

ax0.set_xlim([50, 80])

scatter = ax0.scatter(
    data=over_50,
    x='age', 
    y='chol', 
    c='target',
    cmap="winter"
);

ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=over_50.chol.mean(), linestyle=":");

# Max heart rate plot
ax1.set(
    ylabel='Max heart rate', 
    xlabel='Age',
    title='Maximum heart rate achieved');

ax1.scatter(
    data=over_50, 
    x='age',
    y='thalach',
    c='target',
    cmap="winter"
);

ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
```

```python
fig.savefig("heart-disease-analysis-plot.png")
```
