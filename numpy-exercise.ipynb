```python
np.random.seed(11)
butter_sales = np.random.randint(20, size=(5, 3))
```

```python
butter_sales
```

```python
weekly_sales = pd.DataFrame(data=butter_sales, index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], columns=['Almond Butter', 'Peanut Butter', 'Cashew Butter'])
```

```python
weekly_sales
```

```python
prices = np.array([10, 8, 12])
prices
```

```python
butter_prices = pd.DataFrame(data=prices.reshape(1, 3), index=['Price'], columns=['Almond Butter', 'Peanut Butter', 'Cashew Butter'])
butter_prices
```

```python
weekly_sales['Total'] = butter_prices.dot(weekly_sales.T).T
```

```python
weekly_sales
```

```python
weekly_sales.rename(columns={'Total': 'Total ($)'}, inplace=True)
```
